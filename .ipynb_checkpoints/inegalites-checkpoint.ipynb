```python
import my
import random
```

```python
def remove_edge(v,G,k):
    if G.degree(v)<=k:
        return v
    else:
        n = random.choice(G.neighbors(v))
        G.delete_edge(v,n)
        remove_edge(v, G, k)
```

```python
def min_vertices(G,k):
    vertex_colors = {}
    for v in G.vertices():
        if G.degree(v)==k:
            vertex_colors[v] = 'gray'
        else:
            vertex_colors[v] = 'green'
    return vertex_colors
```

```python
test = kconnected(10,2)
colors = min_vertices(test, 2)
vis = test.plot(vertex_size=200, vertex_colors = my.invert_mapping(colors), title="2-connected graph")
vis.show()
```
