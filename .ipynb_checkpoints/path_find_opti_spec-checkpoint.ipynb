```python
def is_path(path, G):
    for i in range(len(path)-1):
        if not G.has_edge(path[i], path[i+1]):
            return False, i
    return True, None
```

```python
class PartitionJoinError(Exception):
    pass
```

```python
def best_rot(end_access, path, visited):
    for v in end_access:
        tail = my.dfs_pathfinder(G, my.rotation(path, path.index(v)), visited)
        if tail is not None:
            return tail
    raise PartitionJoinError("dead end") 
```

```python
def fix_tail(path, G):
    valid, index = is_path(path, G)
    if valid:
        return path
    else:
        path = path[:index]
        visited = set(path)
        if len(path)<2:
            raise PartitionJoinError("path too short for rotation")
        access = set(G.neighbors(path[-1]))-{path[-2]}
        access = access.intersection(visited)
        if not access:
            raise PartitionJoinError("no accessible vertices")
        unvisited = set(G.vertices())-visited
        end_access = [(v, len(set(G.neighbors(path[path.index(v)+1]))-visited)) for v in access]
        end_access.sort(key=lambda tup: tup[1], reverse=True)
        print("unvisited vertices:", unvisited)
        print("possible rot vertices, sorted:", end_access)
        return best_rot(end_access, path, visited)
```

```python
def fiedler_clusters(fiedler_val, r):
    fiedler_val.sort(key=lambda x: x[1])
    partitions = []
    current_partition = [fiedler_val[0][0]]
    current_base = fiedler_val[0][1]
    
    for v, val in fiedler_val[1:]:
        if val - current_base > radius:
            partitions.append(current_partition)
            current_partition = [v]
            current_base = val
            
```

```python
def fiedler_path_finder(G, k):
    # fiedler et nikiforov : p(G) > n-2 => G Ham
    n = G.order()
    rho = max(G.adjacency_matrix().eigenvalues())
    print("n =", n)
    print("(G) =", rho)
    if rho > n - 2:
        print("(G) > n-2 so graph is likely Hamiltonian")
    else:
        print("spectral condition not met, trying anyway")
        
    L = G.laplacian_matrix()
    eigs = L.eigenvectors_right()
    eigs.sort(key=lambda triple:triple[0])
    if len(eigs)<2:
        raise ValueError("insufficient eigenvectors")
    fiedler_v = eigs[1][1][0]
    
    print("fiedler vector", fiedler_v)
    vertices = G.vertices()
    fiedler_val = {v: fiedler_v[i] for i, v in enumerate(vertices)}
    

    print(vertices)
    fiedler_order = sorted(vertices, key=lambda v:fiedler_val[v])
    
    
    
    
    _path, faulty_index = is_path(fiedler_order, G)
    if _path:
        print("fiedler ordering is already a valid Hamiltonian path")
        return fiedler_order
    else:
        print("gap at index", faulty_index)
        return fix_tail(fiedler_order, G)
```
